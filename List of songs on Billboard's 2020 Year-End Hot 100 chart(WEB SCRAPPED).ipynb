```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_2020'
```

```python
requests.get(url)
```

```python
web = requests.get(url)
```

```python
body = BeautifulSoup(web.text,'html')
body
```

```python
body.find('table')
table = body.find('table')
table
```

```python
table.find_all('th')
head = table.find_all('th')
head
```

```python
[header.text.strip() for header in head]
header = [header.text.strip() for header in head]
header
```

```python
pd.DataFrame(columns=header)
df = pd.DataFrame(columns=header)
df
```

```python
table.find_all('tr')
data = table.find_all('tr')
data
```

```python
#To avoid "row with mismatched columns" we are using the index number on the for loop to remove the empty []
for rows in data[1:]:
    Rows = rows.find_all('td')
    DATA = [Data.text.strip() for Data in Rows]
    length =  len(df)
    df.loc[length] = DATA
df
```
