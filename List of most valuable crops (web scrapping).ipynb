```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most_valuable_crops_and_livestock_products'
```

```python
requests.get(url)
```

```python
web = requests.get(url)
```

```python
body = BeautifulSoup(web.text, 'html')
```

```python
print(body)
```

```python
table = body.find('table')
print(table)
```

```python
titles = table.find_all('th')
print(titles)
```

```python
[clean_title.text.strip() for clean_title in titles]
```

```python
clean_titles = [clean_title.text.strip() for clean_title in titles]
print(clean_titles)
```

```python
pd.DataFrame(columns = clean_titles)
df = pd.DataFrame(columns = clean_titles)
```

```python
column_data = table.find_all('tr')
```

```python
for rows in column_data:
    Rows = rows.find_all('td')
    DATA = [Data.text.strip() for Data in Rows]
    print(DATA)
```

```python
#To avoid "row with mismatched columns" we are using the index number on the for loop to remove the empty []
for rows in column_data[1:]:
    Rows = rows.find_all('td')
    DATA = [Data.text.strip() for Data in Rows]
    length =  len(df)
    df.loc[length] = DATA
df
```
